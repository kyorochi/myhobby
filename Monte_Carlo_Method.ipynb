```python
# -*- coding: utf-8 -*-
import random
import matplotlib.pyplot as plt
import numpy as np 
```

```python
n=5*10**4
np.random.seed(seed=2023)
xy=np.random.uniform(-1, 1, (n, 2))
```

```python
count=0
i_list=np.zeros(n)
pi_list=np.zeros(n)

fig,ax=plt.subplots(figsize=(10,10))

for i in range(n):
    if xy[i,0]**2+xy[i,1]**2 <1.0 :
        count+=1
        ax.plot(xy[i,0],xy[i,1],"ro",markersize=1 ,label="inside the circle")
    else:
       ax.plot(xy[i,0],xy[i,1],"bo",markersize=1,label="outside the circle")
    
    i_list[i]=i
    pi_list[i]=4.0 * count / (i+1)


ax.grid()
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
pi_list_2=np.zeros(n)
for i in range(n):
    pi_list_2[i]=np.pi
```

```python
x=np.linspace(0,n)

fig,ax=plt.subplots(figsize=(12,8))

ax.set_title("Estimation of Pi", fontsize = 16)

ax.set_xlabel("", size = 14)
ax.set_ylabel("", size = 14)

ax.plot(i_list,pi_list,color="b",linestyle = "solid",label="Estimated Pi")
ax.plot(i_list,pi_list_2,color="r",linestyle = "dashed",label="Pi")
ax.legend(prop={"size": 15})
plt.show
print(pi_list[n-1])
```
