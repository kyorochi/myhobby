```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#xのリストを作成
x=np.linspace(-4,4,1000)
#np.linspace(start,stop,要素数)でリストを作成
#フーリエ級数展開（4次まで）
y_1=2*np.sin(x)
y_2=2*np.sin(x)-np.sin(2*x)
y_3=2*np.sin(x)-np.sin(2*x)+2/3*np.sin(3*x)
y_4=2*np.sin(x)-np.sin(2*x)+2/3*np.sin(3*x)-1/2*np.sin(4*x)

#一般のnについても考える
y_n=np.zeros(shape=(x.shape[0]))

for n in range(1,40,1):
    y_n+=2*(-1)**(n+1)/n*np.sin(n*x)

plt.title("f(x) = x foulier expansion")
plt.plot(x,y_1,color="red",linewidth=1.0,label="n=1")
plt.plot(x,y_2,color="blue",linewidth=1.0,label="n=2")
plt.plot(x,y_3,color="purple",linewidth=1.0,label="n=3")
plt.plot(x,y_n,color="gray",linewidth=1.0,label="n=40")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
#グリッドをつける
#画像の保存
plt.savefig("foulier expansion.png")
```
