```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

ここでは,物理学実験2のデータを用いてPythonを使ったデータ解析を行ってみる
.

diode.datを用いる

```python
datax=[0.606100,0.505300,0.402000,0.306000,0.163100,0.038200]
datay=[1.000000,0.100000 ,0.010000 ,0.001000 ,0.000100 ,0.000010 ]
```

```python
print(datax,datay)
```

```python
datax=np.array(datax)
datay=np.array(datay)
```

```python
print(datax,datay)
```

```python
plt.figure(figsize=(12,8),facecolor="white",)
plt.scatter(datax,datay,marker="x",color="blue",linewidth=1.0,label="")
#plt.plot(datax,datay,color="red",linewidth=1.0,label="")

plt.legend()
plt.xlabel("Voltage(Volt)")
plt.ylabel("Current(mA)")
plt.grid()
#plt.xlim()
```

```python
def Exponential_fit(x,a,b):
    return b*(np.exp(a*x)-1)

#curve fittingの実施
param, cov = curve_fit(Exponential_fit, datax, datay)
```

```python
print(param)
print(cov)
```

```python
sample_x=np.linspace(0,0.61,1000)
y=param[1]*(np.exp(param[0]*sample_x)-1)
```

```python
plt.figure(figsize=(12,8),facecolor="white",)
plt.title("deiode")
plt.scatter(datax,datay,marker="+",color="darkviolet",linewidth=1.0,label="Obs.")
plt.plot(sample_x,y,color="mediumspringgreen",linewidth=1.0,label="Calc.")

plt.legend()
plt.xlabel("Voltage(Volt)")
plt.ylabel("Current(mA)")
plt.grid()
plt.xlim(0,0.61)
```

```python
plt.figure(figsize=(12,8),facecolor="white",)
plt.title("diode")
plt.scatter(datax,datay,marker="+",color="darkviolet",linewidth=1.0,label="Obs.")
plt.plot(sample_x,y,color="mediumspringgreen",linewidth=1.0,label="Calc.")

plt.legend()
plt.yscale("log")
plt.xlabel("Voltage(Volt)")
plt.ylabel("Current(mA)")
plt.grid()
plt.ylim(10**(-6),1)
plt.xlim(0,0.61)
```
