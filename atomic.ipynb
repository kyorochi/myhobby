```python
# -*- coding: utf-8 -*-
import random
import matplotlib.pyplot as plt
import numpy as np 
```

```python
t=[1,15,30,45,60,75,90,105,120,135]
N=[106,80,98,75,74,73,49,38,37,22]
sigma_i=[10.3,8.9,9.9,8.7,8.6,8.5,7,6.2,6.1,4.7]

tfit=np.linspace(0,140,1000)
a=-0.009087764739
b=4.72899451
sigma_a=0.00101
sigma_b=0.0649

da1=a+sigma_a
da2=a-sigma_a

db1=b+sigma_b
db2=b-sigma_b

Nfit=np.exp(b)*np.exp(a*tfit)
dy1=np.exp(db1)*np.exp(da1*tfit)
dy2=np.exp(db2)*np.exp(da2*tfit)

t=np.array(t)
N=np.array(N)
sigma_i=np.array(sigma_i)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.set_title("Count of Atomic Decay", fontsize = 16)
ax.set_xlabel("Time[s]", size = 14)
ax.set_ylabel("Count", size = 14)
#ax.scatter(t,N,label="measuring point")
ax.errorbar(t,N,yerr=sigma_i,capsize=5, fmt="x", markersize=10, ecolor='black', markeredgecolor = "red",label="Measuring Point")
ax.plot(tfit,Nfit,"-",color="blue",label="Regression curve")
ax.plot(tfit,dy2,":",color="blue")
ax.plot(tfit,dy1,":",color="blue")
ax.fill_between(tfit,dy2,dy1,color="blue",alpha=0.2,label="Error" )
plt.xlim(0,140)
ax.legend(prop={"size": 15})
```
